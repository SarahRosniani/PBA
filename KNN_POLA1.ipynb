```python
import numpy as np # Digunakan untuk manipulasi array
import pandas as pd # data dalam formata datafram
from matplotlib import pyplot as plt # Untuk visualisai data
from sklearn.model_selection import train_test_split # membagi dataset
from sklearn.metrics import accuracy_score  # mengukur akurasi model
from sklearn.preprocessing import StandardScaler  # penskalaan fitur agar mean 0 dan variansi 1
from sklearn.neighbors import KNeighborsClassifier # membuat/melatih model KNN untuk klasifikais
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/SarahRosniani/Pila/main/Iris%20(1).csv")
```

```python
df.head()
```

```python
df.dropna()
```

```python
X=df.loc[:,'SepalLengthCm':'PetalWidthCm'] #ambil data dri sepat smpe petal buat fitur
y=df.loc[:,'Species'] # label kelas
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
sc = StandardScaler() # penskalaan fitur
X_train = sc.fit_transform(X_train) # untuk data latih mengubah setiap nilai memiliki mean 0 dan devisiasi 1
X_test = sc.transform(X_test) # mengubah data uji yang tlh dihitung sblmnya dari data latih
```

```python
knn=KNeighborsClassifier() # model

# Train the Model
knn.fit(X_train, y_train) # model knn dilatih dengan data latih. menentukan kelas" dri sampel xtrain dan label ytrain
y_pred = knn.predict(X_test) #prediksi data yg telah dilatih pada data uji
```

```python
plt.xlabel('Feature')
plt.ylabel('Species')

X=df.loc[:,'SepalLengthCm']
Y=df.loc[:,'Species']
plt.scatter(X,Y,color='green',label='SepalLengthCm')


X=df.loc[:,'SepalWidthCm']
Y=df.loc[:,'Species']
plt.scatter(X,Y,color='blue',label='SepalWidthCm')


X=df.loc[:,'PetalLengthCm']
Y=df.loc[:,'Species']
plt.scatter(X,Y,color='red',label='PetalLengthCm')


X=df.loc[:,'PetalWidthCm']
Y=df.loc[:,'Species']
plt.scatter(X,Y,color='black',label='PetalWidthCm')

plt.legend(loc=4, prop={'size':5})
plt.show()
```

```python
# hasil akurasi model
accuracy = accuracy_score(y_test, y_pred)*100
print('Accuracy model :' + str(round(accuracy, 2)) + ' %.') # membulatkan nilai akurasi hinga dua desimal
```
